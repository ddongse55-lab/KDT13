### [문제] Seaborn이 제공하고 있는 데이터 셋을 사용하여 질문에 답하시오.

#### (1) 팁 데이터에서 pivot_table()함수를 사용하여 index='day', columns='size', values='tip'으로하는 데이터를 만든 다음 해당 데이터로 히트맵을 작성하시오.

```python
# 팁 데이터 호출
tips = sns.load_dataset('tips')
tips.head()
```

```python
tip_data = tips.pivot_table(index='day', columns='size', values='tip')
tip_data
```

```python
tip_data = tip_data.fillna(0)
tip_data
```

```python
sns.heatmap(tip_data, annot=True, cmap='YlGnBu',
           annot_kws={'size':8, 'weight':'bold', 'color':'black'}, fmt='.2f')
plt.show()
```

#### (2) 아이리스 데이터에서 각 데이터의 상관관계를 pairplot()으로 나타내시오.

```python
# 아이리스 데이터 호출
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, diag_kind='kde', hue='species', markers=['o', 's', 'D'], height=3)
plt.title('Iris Data의 상관관계')
plt.show()
```
