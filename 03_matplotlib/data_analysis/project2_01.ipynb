```python
x = df0['사고유형'].str.contains('차대차', na=False)
y = df0['피해운전자 연령대'] != '12세 이하'

df0[x|y]
```

```python
counts = df0['도로형태'].value_counts()
counts
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 윈도우 폰트 경로 직접 지정 (맑은 고딕)
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_prop = fm.FontProperties(fname=font_path)

# 2. 전체 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False

# 3. 만약 위 방법으로도 제목 등이 안 변한다면, 개별 요소에 직접 적용
# 예: plt.title('제목', fontproperties=font_prop)
```

```python
# 그래프 스타일 설정
sns.set_style("whitegrid") # 그리드가 있는 하얀 배경

# 1. 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 2. 막대그래프 그리기
# x 에는 도로형태(index), y 에는 개수(values)를 넣습니다.
# palette="viridis"는 자동으로 예쁜 색 조합을 적용합니다.
sns.barplot(x=counts.index, y=counts.values, palette="viridis")

# 3. 꾸미기
plt.title('도로형태별 사고 건수', fontsize=16, fontweight='bold', fontproperties=font_prop)
plt.xlabel('도로형태', fontsize=12, fontproperties=font_prop)
plt.ylabel('사고 건수', fontsize=12, fontproperties=font_prop)

# x축 글씨 회전 
plt.xticks(rotation=45, ha='right', fontproperties=font_prop)

# 4. 출력
plt.tight_layout()
plt.show()
```
