```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

import matplotlib.font_manager as fm

# 폰트 경로 지정 (윈도우의 맑은 고딕 경로)
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
df1 = pd.read_csv('./accident_2013_2015.csv')
df1
```

```python
df2 = pd.read_csv('./accident_2016_2018.csv')
df2
```

```python
df3 = pd.read_csv('./accident_2019_2021.csv')
df3
```

```python
df4 = pd.read_csv('./accident_2022_2024.csv')
df4
```

```python
df0 = pd.concat([df1, df2, df3, df4], ignore_index=True)
df0
```

```python
x = df0['사고유형'].str.contains('차대차', na=False)
y = df0['피해운전자 연령대'] != '12세 이하'

df0[x|y]
```

```python
condition = x|y

df0 = df0[~condition].reset_index(drop=True)
df0
```

```python
x = df0['사고유형'].str.contains('차대차', na=False)
y = df0['피해운전자 연령대'] != '12세 이하'

df0[x|y]
```

```python
df_sum = df0.groupby('도로형태')[['사망자수', '중상자수', '경상자수']].sum()
df_sum
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 1. 윈도우 폰트 경로 직접 지정 (맑은 고딕)
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_prop = fm.FontProperties(fname=font_path)

# 2. 전체 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())
plt.rcParams['axes.unicode_minus'] = False

# 3. 만약 위 방법으로도 제목 등이 안 변한다면, 개별 요소에 직접 적용
# 예: plt.title('제목', fontproperties=font_prop)
```

```python
# 3. 그래프 그리기
ax = df_sum.plot(kind='bar', figsize=(12, 6), rot=45)

# 4. 한글 타이틀 및 레이블 설정
plt.title('도로형태별 사상자 수 비교', fontsize=16, fontweight='bold', fontproperties=font_prop)
plt.xlabel('도로형태', fontsize=12, fontproperties=font_prop)
plt.ylabel('인원 수', fontsize=12, fontproperties=font_prop)
plt.legend(title='사상자 구분')

# 5. 출력
plt.tight_layout()
plt.show()
```
