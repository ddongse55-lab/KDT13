## 라인 플롯(line plot)

### figure 객체
#### axes : 차트를 그릴 수 있는 영역
#### axis : 차트의 축

```python
import matplotlib.pyplot as plt

plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축)
# 'bo--' : b -> 선 색상, o -> 마커, -- -> 선 종류
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--')
plt.show
```

```python
plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축)
#          (x축 정보만 제공한 경우, y축은 자동 생성)
plt.plot([10, 20, 30, 40])
plt.show
```

```python
plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축, 스타일)
#          (스타일 지정안할시 default 실선으로 그려짐)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show
```

```python
import numpy as np

x = np.arange(-4.5, 5, 0.5)
y = 2*x**2
plt.plot(x,y)
plt.show
```

### subplot() : 1개의 figure에 여러 개의 차트를 그리는 함수

```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10

plt.plot(x,y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show
```

```python
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2

# figure() : figure 객체자 존재하지 않을 경우 생성 함수
#            이미 존재하는 경우 객체를 활성화하는 함수
#            () 안의 내용은 figure를 구분하기 위한 일종의 식별자라고 생각하면 됨
plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y2)

plt.figure(2)
# clf() : figure 객체릴 clear하는 함수
plt.clf()

plt.show
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.title('라인 플롯(line plot)')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0, 50)
plt.ylim(0, 20)
plt.show
```

```python
font_location = 'C:\Windows\Fonts\malgun.ttf'
font_name = font_manager.FontProperties(fname=font_location).get_name()
plt.rc('font', family=font_name)
plt.title('라인 플롯(line plot)', fontdict=font)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0, 50)
plt.ylim(0, 20)
plt.savefig('fig01.png', dpi=100)
plt.show
```

```python
# linspace(-np.pi, np.pi, 256) : 시작부터 끝까지 256개의 값을 만듦
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title('LaTex, 문자열로 tick label 정의')
plt.plot(X, C)
plt.xticks([-np.pi, np.pi / 2, 0, np.pi / 2, np.pi],
          [r'$-/pi$', r'$-/pi2$', r'$0$', r'$+/pi2$', r'$+/pi$']) # 라벨명 -> Latex 수학문자식 
plt.yticks([-1, 0, 1], ['Low', 'Zero', 'High'])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2.2, 0.2),
           labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2.2, 0.2),
           labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))
plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')
plt.show()
```

### tick_params() :  눈금의 스타일을 다양하게 설정
- axis: 설정이 적용될 축을 지정
- direction: 'in'/'out'으로 설정하면 눈금이 안/밖으로 표시 {'in'/'out'/'inout' 中 선택}
- length: 눈금의 길이를 지정
- pad: 눈금과 레이블의 거리를 지정
- labelsize: 레이블의 크기를 지정
- labelcolor: 레이블의 색상을 지정
- top/bottom/left/right: True/False로 지정하면 눈금이 표시될 위치를 선택
- width: 눈금의 너비를 지정
- color: 눈금의 색상을 지정

```python
t = np.arange(0., 5., 0.2)
plt.title('Line Plot')
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.legend(['Red', 'Blue', 'Green'], loc = 'best', ncol=3, fontsize=11, frameon=False, shadow=True)  # ncol=3 : 범례의 열 개수 지정
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.title('Line Plot')
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot( t, 0.2 * t**3, 'g^-')
plt.legend(['Red', 'Blue', 'Green'], loc = 'best', ncol=3, fontsize=11, frameon=False, shadow=True)  # ncol=3 : 범례의 열 개수 지정
plt.show()
```
