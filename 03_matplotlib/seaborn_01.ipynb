```python
# load_dataset : seaborn이 정의하고 있는 데이터를 가지고 올 때 사용
tips = sns.load_dataset('tips')
tips.head()
```

```python
import matplotlib.pyplot as plt

sns.set_style('white')        # darkgrid, whitegrid, ticks, dark, white
sns.set_context('notebook')   # paper, notebook, talk, poster
sns.set_palette('bright')     # deep, muted, pastel, dark, bright, colorblind
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
plt.show()
```

### joint plot

```python
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde', color='g')
plt.title('꽃받침의 길이와 넓이의 Joint Plot', y=1.02)
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter')
plt.title('꽃받침의 길이와 넓이의 Joint Plot', y=1.02)
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex', color='g')
plt.title('꽃받침의 길이와 넓이의 Joint Plot', y=1.02)
plt.show()
```

### Count Plot

```python
titanic = sns.load_dataset('titanic')
sns.countplot(y='class', data=titanic, hue='who')
plt.title('타이타닉호의 각 클래스 별 승객 수')
plt.show()
```

### Pair Plot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, diag_kind='kde', hue='species', markers=['o', 's', 'D'], height=3)
plt.title('Iris Data의 Pair Plot')
plt.show()
```

### Heat Map

```python
titanic = sns.load_dataset('titanic')
titanic.head(2)
```

```python
titanic_size = titanic.pivot_table(index='class', columns='sex', aggfunc='size')
titanic_size
```

```python
# annot=True : 값을 출력
# fmt='d' : 정수만 사용(format 형식)
sns.heatmap(titanic_size, cmap=sns.light_palette('gray', as_cmap=True), annot=True, fmt='d')
plt.title('Heatmap')
plt.show()
```

```python
titanic.corr(numeric_only=True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='YlGnBu',
           annot_kws={'size':8, 'weight':'bold', 'color':'black'}, fmt='.2f')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.barplot(x='day', y='total_bill', data=tips, hue='sex')
plt.title('요일별 전체 팁')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.boxplot(data=tips, orient='v')
plt.title('요일 별 전체 팀의 Box Plot')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('요일 별 전체 팁의 Violin Plot')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.title('요일 별 전체 팁의 Strip Plot')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('요일 별 전체 팁의 Strip Plot')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
data= titanic[titanic.survived.notnull()]
sns.catplot(x="age", y="sex", hue="survived", row="class", data=data,
            kind="violin", split=True, height=2, aspect=4)

plt.title("각 클래스의 성별 생존자 수의 Catplot")
plt.show()
```

```python
sns.catplot(x='who', y='survived',
       col='pclass',
       row='embarked',
       kind='bar',
       height=2.5,
       data=titanic)
plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.title("Boxplot과 Strip Plot로 표현한, 요일 별 팁")

sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4")

plt.show()
```
