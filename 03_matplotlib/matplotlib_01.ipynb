## 라인 플롯(line plot)

### figure 객체
#### axes : 차트를 그릴 수 있는 영역
#### axis : 차트의 축

```python
import matplotlib.pyplot as plt

plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축)
# 'bo--' : b -> 선 색상, o -> 마커, -- -> 선 종류
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'bo--')
plt.show()
```

```python
plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축)
#          (x축 정보만 제공한 경우, y축은 자동 생성)
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.title('line plot')
# plot() : line 차트를 그리는 함수 
#          (왼쪽 - x축, 오른쪽 - y축, 스타일)
#          (스타일 지정안할시 default 실선으로 그려짐)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

### subplot() : 1개의 figure에 여러 개의 차트를 그리는 함수

```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10

plt.plot(x,y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2

# figure() : figure 객체자 존재하지 않을 경우 생성 함수
#            이미 존재하는 경우 객체를 활성화하는 함수
#            () 안의 내용은 figure를 구분하기 위한 일종의 식별자라고 생각하면 됨
plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y2)

plt.figure(2)
# clf() : figure 객체릴 clear하는 함수
plt.clf()

plt.show()
```

### tick_params() :  눈금의 스타일을 다양하게 설정
- axis: 설정이 적용될 축을 지정
- direction: 'in'/'out'으로 설정하면 눈금이 안/밖으로 표시 {'in'/'out'/'inout' 中 선택}
- length: 눈금의 길이를 지정
- pad: 눈금과 레이블의 거리를 지정
- labelsize: 레이블의 크기를 지정
- labelcolor: 레이블의 색상을 지정
- top/bottom/left/right: True/False로 지정하면 눈금이 표시될 위치를 선택
- width: 눈금의 너비를 지정
- color: 눈금의 색상을 지정

```python
t = np.arange(0., 5., 0.2)
plt.title('Line Plot')
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.legend(['Red', 'Blue', 'Green'], loc = 'best', ncol=3, fontsize=11, frameon=False, shadow=True)  # ncol=3 : 범례의 열 개수 지정
plt.show()
```

```python
mpl.rcParams['axes.unicode_minus'] = False
x1 = np.linspace(start=0.0, stop=5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)
y1 = np.cos(2 * np.pi * x1) *np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle('Main Title')

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2-1 subplots')
plt.xlabel('time (m)')
plt.ylabel('damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('A tale of 2-2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(- x**2 - y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
plt.title('Contour Plots')
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='hot')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = [70, 90, 80]
y2 = [60, 70, 90]

x = np.arange(len(y1))

err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

xlabel = ['국어', '영어', '수학']

plt.title("Bar Chart")

plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center', label="점수")
plt.bar(x+0.3, y2, yerr=err2, width=0.3, color='blue', align='center', label="반평균")

plt.xticks(x+0.3/2, xlabel)
plt.ylim(0, 100)

plt.ylabel("점수")
plt.xlabel("과목")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = [70, 90, 80]
y2 = [60, 70, 90]

x = np.arange(len(y1))

err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5

ylabel = ['국어', '영어', '수학']

plt.title("Barh Chart")

plt.barh(x, y1, height=0.3, xerr=err1, ecolor='yellow', color='red', align='center', label="점수")
plt.barh(x+0.3, y2, height=0.3,xerr=err2, color='blue', align='center', label="반평균")

plt.yticks(x+0.3/2, ylabel)
plt.xlim(0, 100)

plt.xlabel("점수")
plt.ylabel("과목")

plt.legend()
plt.show()
```

```python
import pandas as pd

plt.rc('font', family='Malgun Gothic')
plt.rc('font', size=12)
plt.rc('axes', unicode_minus=False)

year = [2018, 2019, 2020, 2021, 2022]

physics = [6.2, 6.3, 6.0, 5.9, 6.8]
chemistry = [10.6, 9.4, 8.0, 8.0, 8.0]
life = [16.1, 16.3, 13.9, 13.1, 14.7]
earth = [16.7, 17.7, 16.1, 13.0, 14.9]

science = {
    'year': year,
    'physics': physics,
    'chemistry': chemistry,
    'life': life,
    'earth': earth
}

sci = pd.DataFrame(science)
sci
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.bar(sci['year'], sci['life'], bottom=sci['earth'])
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(sci['year'], sci['earth'])
plt.bar(sci['year'], sci['life'], bottom=sci['earth'])
plt.bar(sci['year'], sci['chemistry'], bottom=sci['earth'] + sci['life'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('학년도 기준 수능 과학탐구 응시자 수')
plt.xlabel('학년도')
plt.ylabel('응시자 수(단위 : 만 명)')

plt.bar(sci['year'], sci['earth'], label='지구과학')
plt.bar(sci['year'], sci['life'], bottom=sci['earth'], label='생명과학')
plt.bar(sci['year'], sci['chemistry'], bottom=sci['earth'] + sci['life'], label='화학')
plt.bar(sci['year'], sci['physics'], 
        bottom=sci['earth'] + sci['life'] + sci['chemistry'], 
        label='물리학')

plt.legend(ncol=2)
plt.show()
```

## 파이 차트(pie chart)

```python
labels = ['개구리', '돼지', '개', '동나무']
sizes = [15.55, 30, 45, 9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title("Pie Chart")

plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.2f%%',
    shadow=True,
    startangle=90,
    counterclock=False
)

plt.axis('equal')
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(
    ratio,
    labels=labels,
    autopct='%.1f%%',
    startangle=260,
    counterclock=False,
    colors=colors,
    wedgeprops=wedgeprops
)

plt.show()
```
