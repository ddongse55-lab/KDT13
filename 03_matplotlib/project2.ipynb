```python
region = ['서울', '대전', '대구', '부산', '광주', '경기', '강원', '경상', '전라', '충청', '제주']
num = []

for i in region:
    data = data1['제공기관명'].str.contains(i).mean()
    num.append(data)

n_data = len(region)
index = np.arange(n_data)
plt.bar(index, num, tick_label=region)
plt.show()
```

```python
counts = df0['도로형태'].value_counts()
counts
```

```python
# 그래프 스타일 설정
sns.set_style("whitegrid") # 그리드가 있는 하얀 배경

# 1. 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 2. 막대그래프 그리기
# x 에는 도로형태(index), y 에는 개수(values)를 넣습니다.
# palette="viridis"는 자동으로 예쁜 색 조합을 적용합니다.
sns.barplot(x=counts.index, y=counts.values, palette="viridis")

# 3. 꾸미기
plt.title('도로형태별 사고 건수', fontsize=16, fontweight='bold')
plt.xlabel('도로형태', fontsize=12)
plt.ylabel('사고 건수', fontsize=12)

# x축 글씨 회전 
plt.xticks(rotation=45, ha='right')

# 4. 출력
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. 그래프 크기 및 레이아웃 설정
plt.figure(figsize=(12, 6)) # 그래프의 가로, 세로 크기 설정 (데이터가 많아 가로로 길게)

# 2. 막대그래프 그리기
# counts.index는 도로형태 이름, counts.values는 개수입니다.
plt.bar(counts.index, counts.values)

# 3. 꾸미기
plt.title('도로형태별 사고 건수', fontsize=16, fontweight='bold') # 제목
plt.xlabel('도로형태', fontsize=12)             # x축 이름
plt.ylabel('사고 건수', fontsize=12)             # y축 이름

# x축 글씨가 겹칠 때: 45도 회전시키기 (강력 추천)
plt.xticks(rotation=45, ha='right')

# 그리드(눈금선) 추가로 보기 편하게 만들기
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 4. 출력
plt.tight_layout() # 글씨가 잘리지 않게 레이아웃 자동 조정
plt.show()
```
