```python
from pandas import Series

data = [1000, 2000, 3000]
s = Series(data)
print(s.index)
# index.to_list 메서드를 사용하면 인덱스를 리스트 형태로 출력 가능
print(s.index.to_list())
```

```python
data = [1000, 2000, 3000]
# 각 데이터에 정수 인덱스가 자동으로 부여
s = Series(data)
# 인덱스를 입력한 값으로 변경
s.index = ['메로나', '구구콘', '하겐다즈']
print(s)
```

```python
data = [1000, 2000, 3000]
index = ['메로나', '구구콘', '하겐다즈']
# s = Series(data=data, index=index)
# s = Series(data, index=index)
# s = Series(index=index, data=data) 모두 실행 가능
# s = Series(data=data, index) 일반매개변수 앞에 키워드매개변수가 올 수 없으므로 실행 불가능
s = Series(data, index)
print(s)
```

```python
# reindex 메서드 : 인덱스의 이름을 재설정 
#                  -> 이전에 없던 인덱스 생성 시 자료값이 None으로 생성되어 삽입
#                  -> 이 경우 dtype이 float으로 변경됨
s2 = s.reindex(['메로나', '비비빅', '구구콘'])
print(s2)
```

```python
# fillna 메서드 : 자료값이 None인 인덱스를 찾아 자료값을 0으로 채운다
s2 = s2.fillna(0)
print(s2)
```

```python
price = [42500, 42550, 41800, 42550, 42650]
# date와 같이 인덱스 값에 연,월,일 등 날짜 또는 시간 형태의 값이 입력되면 이것을 시계열 데이터라고 함
date = ['2019-05-31', '2019-05-30', '2019-05-29', '2019-05-28', '2019-05-27']
s = Series(price, date)
print(s)
```

## Series indexing

```python
s1 = Series([10, 20, 30])
s2 = Series([10, 20, 30], index=[1, 2, 3])

print(s1[0])
# print(s2[0]) : 존재하지 않는 인덱스로 접근하기 때문에 에러 발생
print(s2[1])
```

```python
s3 = Series([10, 20, 30], index=['a', 'b', 'c'])
# 딕셔너리에 접근하듯이 key-value 형식으로 접근
print(s3['a'])
```

```python
print(s3.index[0])
print(s3.index[-1])
```

## Series 연산

```python
s1 = pd.Series([9904312, 3448737, 2890451, 2466052],
               index=['서울', '부산', '인천', '대구'])
s2 = pd.Series({'서울' : 9631482, '부산' : 3393191, '인천' : 2632035, '대전' : 1490158},
               index=['부산', '서울', '인천', '대전'])
ds = s1 - s2
print(ds)
```

```python
# 브로드캐스팅 연산 : 모든 값들에 연산이 각각 적용
s3 = s1 / 1000000
print(s3)
```

```python
# 값만 가져와서 같은 자리에 있는 값 끼리 연산 수행
print(s1.values - s2.values)
```

```python
# notnull() : 자료값이 null이 아닌 부분(T/F) -> 데이터 필터링에 사용
# ds[ds.notnull()] : notnull에서 True인 부분만 추출
print(ds[ds.notnull()])
```

```python
rs = (s1 - s2) / s2 * 100
print(rs)
```

```python
rs = rs[rs.notnull()]
print(rs)
```

```python
rs['부산'] = 1.63     # 기존에 존재하던 값이 있으므로 값을 수정
rs['대구'] = 1.41     # 기존에 존재하던 값이 없으므로 새로운 값을 추가
print(rs)
```

```python
del rs['서울']       # 기존에 존재하던 값을 삭제
print(rs)
```

```python
from pandas import Series

date = ['6/1', '6/2', '6/3', '6/4', '6/5']
high = Series([42800, 42700, 42050, 42950, 43000], index=date)
low = Series([42150, 42150, 41300, 42150, 42350], index = date)
print(high)
print('------------')
print(low)
```

```python
diff = high - low
print(diff)
```

```python
# 가장 큰 값
print(diff.max())
# 가장 큰 값의 위치
print(diff.idxmax())
```

```python
def remove_comma(x):
    # replace(',', '') : ','를 ''으로 대체
    return int(x.replace(',', ''))
```

```python
s = Series(['1,234', '5,678', '9,876'])
# pandas에서 map 함수는 반복자 객체가 따로 주어지지 않고, Series가 가진 객체의 수 만큼 객체를 차례로 호출하게 됨
result = list(s.map(remove_comma))
print(result)
```

```python
from pandas import Series

data = [3.1, 2.0, 10.1, 5.1]
index = ['00010', '00020', '00030', '00040']
s = Series(data, index)
print(s)
```

```python
# Series 객체의 자료값을 기준으로 오름차순으로 정렬
# ascending=False : 내림차순으로 정렬
s1 = s.sort_values(ascending = False)
print(s1)
```

```python
# 자료 값(원데이타)을 기준으로 인덱스 값을 오름차순으로 정렬
# ascending=True : 기본값
# ascending=False : 내림차순으로 정렬
s2 = s.rank()
print(s2)
```

```python
s2 = s.rank(ascending=False)
print(s2)
```
