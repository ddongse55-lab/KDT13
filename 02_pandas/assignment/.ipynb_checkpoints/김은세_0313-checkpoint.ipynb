```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris = sns.load_dataset('iris')
iris
```

## [문제1]

##### (1) iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.

```python
iris_mean = iris.groupby('species')[['sepal_length']].mean()
iris_mean
```

```python
iris_mean.plot.bar(rot=0)
plt.title('Iris 품종 별 꽃받침 길이의 평균')
plt.xlabel('Iris 품종')
plt.ylabel('꽃받침의 길이')
plt.show()
```

##### (2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.

```python
iris_size = iris.groupby('species').size()
iris_size
```

```python
iris_size.plot.pie(autopct='%.2f%%')
plt.title('Iris 품종 별 데이터의 개수')
plt.axis('equal')
plt.show()
```
