## Titanic 예제 실습

```python
import pandas as pd
```

```python
# read_csv() : csv 파일을 읽어오는 메서드
# train Data : 학습이 완료된 데이터
# test Data : 학습이 올바르게 완료된 것인지 확인하기 위한 데이터
df = pd.read_csv('./titanic/train.csv')
df
```

### column별 의미
    - PassengerId : 각 승객에게 주어진 고유 ID 번호
    - Survived : 생존여부(0 = 사망, 1 = 생존)
                 -> Target Data(= 종속 변수)
    - Pclass : 티켓 클래스(1 = 1등석, 2 = 2등석, 3 = 3등석) 
    - Sex : 성별
    - Age : 나이
    - SibSp : 함께 탑승한 형제자매/배우자의 수
    - Parch : 함께 탑승한 부모님/아이들의 수
    - Ticket : 티켓 번호
    - Fare : 탑승 요금
    - Cabin : 수하물 번호 또는 객실 카테고리
    - Embarked : 선착장
                 (C = Cherboug, Q = Queenstown, S = Southampton)

```python
def n_square(x):
    return x**2
```

```python
def n_square(x,n):
    return x**n
```

## 객체 연산

```python
data = [[1, 10, 100], [2, 20, 200], [3, 30, 300]]
col = ['col1', 'col2', 'col3']
row = ['row1', 'row2', 'row3']
df = pd.DataFrame(data, row, col)
df
```

```python
result = df.add(1)
result
```

```python
result = df + 1
result
```

```python
data2 = [[3], [4], [5]]
df2 = pd.DataFrame(data2, index=['row1', 'row2', 'row3'], columns=['col1'])
df2
```

```python
# add() : 덧셈
# sub() : 뺄셈
# mul() : 곱셈
# div() : 나눗셈
# mod() : 나머지 연산
result = df.add(df2)
result
```

```python
# fill_value(0) : 동일한 부분이 없는 경우에, 0으로 채워서 연산을 진행
result = df.add(df2, fill_value=0)
result
```

```python
period = pd.period_range(start='2022-01-13 00:00:00', end='2022-01-13 02:30:00', freq='30min')
data = {'col1':[1,2,3,4,5,6], 'col2':period}
idx = ['row1', 'row2', 'row3', 'row4', 'row5', 'row6']
df = pd.DataFrame(data, idx)
df
```

```python
# rolling() : 기간을 주고 데이터를 구하는 것
#             ex) rolling(window=3).sum() : 3개의 데이터를 가지고 누적된 합을  구하는 것
df.rolling(window=3).sum(numeric_only=True)
```

```python
# rolling(window=3).mean() : 3개의 데이터를 가지고 누적된 평균을  구하는 것
df.rolling(window=3).mean(numeric_only=True)
```
