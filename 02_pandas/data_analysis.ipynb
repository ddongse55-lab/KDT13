```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 불러오기
df = pd.read_csv('./project1/pohang_train_2018_2023.csv')
df
```

### column 별 의미
- datetime : 강수량 측정 날짜 및 시간
- hourly_rain : 시간별 강수량
- daily_rain : 일별 강수량
- water_level : 형산강의 일별 수위
- heavy_rain : 호우여부(호우 = 1, 호우X = 0, 호우 경보의 유무로 기준)

```python
# 한글로 되어있는 인덱스 영어로 바꾸기
df1 = df.rename(columns = {
    '일시' : 'datetime',
    '강수량(mm)' : 'hourly_rain'})
df1
```

```python
# 'datetime'을 연'year'/월'month'/일'day'/시'time'로 구분해서 새로운 clumn 만들기
df1['datetime'] = pd.to_datetime(df1['datetime'])
df1['year'] = df1['datetime'].dt.year
df1['month'] = df1['datetime'].dt.month
df1['day'] = df1['datetime'].dt.day
df1['hour'] = df1['datetime'].dt.hour
df1
```

```python
# 필요한 column들만 모으기
df1 = df[['year', 'month', 'day', 'hour', 'hourly_rain', 'daily_rain', 'water_level', 'heavy_rain']]
df1
```

### 누적 강수량 Feature 만들기

```python
# 3시간 누적 강수량
df1['rain_3h'] = df1['hourly_rain'].rolling(3).sum()
# 6시간 누적 강수량
df1['rain_6h'] = df1['hourly_rain'].rolling(6).sum()
# 9시간 누적 강수량
df1['rain_9h'] = df1['hourly_rain'].rolling(9).sum()
# 12시간 누적 강수량
df1['rain_12h'] = df1['hourly_rain'].rolling(12).sum()
# Nan 데이터 0으로 채우기
df1 = df1.fillna(0)
df1
```
