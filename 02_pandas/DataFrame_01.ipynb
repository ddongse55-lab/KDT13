### column별 의미
    - PassengerId : 각 승객에게 주어진 고유 ID 번호
    - Survived : 생존여부(0 = 사망, 1 = 생존)
                 -> Target Data(= 종속 변수)
    - Pclass : 티켓 클래스(1 = 1등석, 2 = 2등석, 3 = 3등석) 
    - Sex : 성별
    - Age : 나이
    - SibSp : 함께 탑승한 형제자매/배우자의 수
    - Parch : 함께 탑승한 부모님/아이들의 수
    - Ticket : 티켓 번호
    - Fare : 탑승 요금
    - Cabin : 수하물 번호 또는 객실 카테고리
    - Embarked : 선착장
                 (C = Cherboug, Q = Queenstown, S = Southampton)

### merge()

```python
df1 = pd.DataFrame({'고객번호':[1001, 1002, 1003, 1004, 1005, 1006, 1007],
                   '이름':['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']},
                  columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({'고객번호':[1001, 1001, 1005, 1006, 1008, 1001],
                   '금액':[10000, 20000, 15000, 5000, 100000, 30000]},
                  columns=['고객번호', '금액'])
df2
```

```python
# 기준 column을 따로 정하지않으면 공통 column으로 병합을 진행
df3 = pd.merge(df1, df2)
df3
```

```python
df4 = pd.merge(df1, df2, how='outer')
df4
```

```python
# 왼쪽(df1)에 위치한 항목은 중복되더라도 추가
df5 = pd.merge(df1, df2, how='left')
df5
```

```python
# 오른쪽(df2)에 위치한 항목은 중복되더라도 추가
df6 = pd.merge(df1, df2, how='right')
df6
```
